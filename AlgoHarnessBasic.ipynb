```python
# Classification Example w/ builtin breast cancer dataset
```

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
data = load_breast_cancer()
```

```python
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = .5,random_state = 21)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
```

```python
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```

```python
print(models)
```

```python
from lazypredict.Supervised import LazyRegressor
from sklearn import datasets
from sklearn.utils import shuffle
import numpy as np
```

```python
# load the toy dataset (boston)
boston = datasets.load_boston()
```

```python
X,y = shuffle(boston.data, boston.target, random_state = 21)
X = X.astype(np.float32)
```

```python
offset = int(X.shape[0] * 0.9)
```

```python
X_train, y_train = X[:offset],y[:offset]
```

```python
X_test, y_test = X[offset:],y[offset:]
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
```

```python
models, predictions = reg.fit(X_train, X_test, y_train, y_test)
```

```python
print(models)
```
